```python
# Grafico 1
# Eliminiar los duplicados
athletes_medals_dataset = athletes_medals_dataset.drop_duplicates(subset=['Year','Name'])

# AÃ±os
all_years = athletes_medals_dataset['Year']
print(all_years)
print('----------------------------------------------------')
years = pd.unique(all_years)
print(len(years))

# Deportes
all_sports = athletes_medals_dataset['Sport']
print(all_sports)
print('----------------------------------------------------')
sports = pd.unique(all_sports)
print(sports)
print('----------------------------------------------------')
print(len(sports))

sports_pivot = athletes_medals_dataset.pivot_table('Age', index='Sport', columns='Year', aggfunc='count')
display(sports_pivot)

plot = pd.DataFrame(sports_pivot.count(), columns = ['Number of sports'])
plot['Year'] = plot.index

display(plot)

years_array = plot['Year'].to_list()
# display(years_array)

sports_array = plot['Number of sports'].to_list()
# display(sports_array)

# Difuminado de colores utilizando modelo HSB Matiz, SaturaciÃ³n y Brillo
data_length = len(years_array)
cmap = plt.cm.get_cmap('hsv',data_length)
cmap_array = [cmap(i) for i in range(data_length)]

fig, ax = plt.subplots(figsize=(10,20))
width = 0.8
opacity = 0.7
rects = ax.bar(years_array, sports_array, width, alpha=opacity, color=cmap_array)

## TÃ­tulo de GrÃ¡fica
plt.title('Number of sports over the years')
## Leyenda en el eje y
plt.ylabel('Number of sports')
## Leyenda en el eje x
plt.xlabel('Years')


plt.show()
```
